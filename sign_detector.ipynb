```python
import cv2, time
#import cv
from datetime import datetime
import datetime
video = cv2.VideoCapture(0)
```

```python
while True:
    check,frame = video.read()#opening camera and taking input
    #print(check)
    #print(frame)
    #cv2.imshow("Video",frame)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #region of interest
    area = gray[100:400,320:620]
    cv2.imshow('area', area) 

    image = cv2.resize(area, (28,28), interpolation = cv2.INTER_AREA)
    resized_image=image/255.0    
    cv2.imshow('image rescale', resized_image)
    copy = frame.copy()
    cv2.rectangle(copy,(320,100), (620,400), color=(0,0,0), thickness=5)
    image = resized_image.reshape(1,28,28,1)
    
    
    result = model.predict_classes(image, 1)[0]
    cv2.putText(copy, str(classLabels[result]), org, font,fontScale, color, thickness, cv2.LINE_AA)
    print(str(classLabels[result]))
    cv2.imshow("output",copy)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
            
            
            
            
            
video.release()
cv2.destroyAllWindows()
```
